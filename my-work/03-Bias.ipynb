```python
# Generate a sanple of values - note we see the population standard deviation
x = np.random.normal(10.0, 1.0, 1000)
x
```

Now let us calculate the mean and investigate it

```python
# Calculate the mean
x_mean = x.mean()

zeroed = x - x_mean

# have alook at the zeroed values
zeroed
```

```python
# What do you yhink the mean of zeroed is ?
zeroed.mean()
```

```python
# This will give us a better view of it - correct to four decimal places
print(f'{zeroed.mean():0.4f}')
```

So, subtraction the mean of the sample results in the mean being zero

The goal is to summarise the zeroed array without losing infoemation about the distance of each point from the mean

Plot idea

```python
# Create a plot
fig, ax = plt.subplots(figsize=(12, 6))

# plot the zeroed array each value spaced out evenly along x axis.
# note the x axis is just the position of the value in the zeroed array
ax.plot(range(len(zeroed)), zeroed, 'k.')

# plot the y=0 line
ax.axhline(y=0.0, color='grey', linestyle='-');
```

One idea is to take the average (vertical) distance each point is from the mean, zero

We need to be careful - theare are negative and positive values here

By definition they sum to zero

```python
# sum the array
print(f'{zeroed.sum():0.4f}')
```

We can try taking the absolute value

```python
# Absolute values.
np.abs(zeroed)
```

```python
# Average absolute value
np. mean(np.abs(zeroed))
```

```python
# Square the values
np.square(zeroed)
```

Note that when you square numbers, bigger values get bigger

Larger deviation from the mean will contribute relatively more to the standard deviation

This is not necassarilly a bad thing it is just a note

```python
# Create a plot
fig, ax = plt.subplots(figsize=(12, 6))

# plot the squard zeroed array each value spaced out evenly along x axis.
# note the x axis is just the position of the value in the zeroed array
ax.plot(range(len(zeroed)), np.square(zeroed), color='green', marker='.', linestyle='none')

# plot the zeroed array each value spaced out evenly along x axis.
# note the x axis is just the position of the value in the zeroed array
ax.plot(range(len(zeroed)), zeroed, 'k.')

# plot the y=0 line
ax.axhline(y=0.0, color='grey', linestyle='-');
```

```python
# Calculate the average squared result
np.mean(np.square(zeroed))
```

Now because we have squared the original values we often take the square root of the answer

```python
# Calculate the square root of the average squared result
np.sqrt(np.mean(np.square(zeroed)))
```

### Bassel's Correction

We will use small samples as the effect is clearer

```python
# Create 100000 samples of size 5 - standard deviation is 2.0
samples = np.random.normal(0.0, 2.0, (100000, 5))
samples
```

```python
# Calculate standard deviation without correction
stdev = samples.std(axis=1)
stdev
```

```python
# View a histogram - hopfully we can see the extimate is too small
fig, ax = plt.subplots(figsize=(12, 6))

# Plot histogram
plt.hist(stdev, bins=100)

# Draw a verical line where the actual standard devistion is
plt.axvline(x=2, color='red');
```

It seems clear that the tip of the curve is below the actual value

### The Correction

Bassel's correction actually applies to the variance

The variance of a sample is the square of the standard deviation

It is what we get is we do not apply the np.sqrt function in the standard deviation calculation

The correction is to multiply the calculation be n/(n-1)

```python
# Uncorrected variance
np.mean(np.square(x-np.mean(x)))
```

```python
# Corrected variance
np.mean(np.square(x- np.mean(x))) * (len(x) / (len(x) -1.0))
```

The correction can eb applied to the standard deviation directly but will still lead to a somewhat biased estimator
